```python
import pandas as pd

# Load simulated price-revenue data
df_sim = pd.read_csv('../data/price_simulation_results.csv')
df_sim.head()
```

```python
# Get the row with maximum revenue for each StockCode
df_optimal_price = df_sim.loc[df_sim.groupby('StockCode')['SimulatedRevenue'].idxmax()].reset_index(drop=True)

# Optional: Sort and preview
df_optimal_price = df_optimal_price.sort_values(by='SimulatedRevenue', ascending=False)
df_optimal_price.head()
```

```python
df_optimal_price.to_csv('../data/optimal_price_table.csv', index=False)
```

Optimal Price Selection Summary

We used the simulated demand and revenue data  to determine the best price point for each product.

For every `StockCode`, we selected the price that resulted in the maximum revenue. This gives us a practical, data-driven pricing strategy tailored for each item.

This table will now be used in our business summary and the Streamlit dashboard for actionable insights.
